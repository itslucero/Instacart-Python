### 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Check frequency of column 'order_hour_of_day'

ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

There are 24 hours in a day, will divide the counts by 3 "most orders", "average orders" and "fewest orders"

```python
# create a list for percentiles

list_quantile = list(ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).quantile([0.75, 0.5, 0.25]))
```

```python
list_quantile
```

```python
dict_hour_counts = ords_prods_merged['order_hour_of_day'].value_counts(dropna=False).to_dict()
```

```python
result = []

for value in ords_prods_merged['order_hour_of_day']:
    for key, val in dict_hour_counts.items():
        if value == key:
            if val < list_quantile[2]:
                result.append('Fewest Orders')
            elif val >= list_quantile[2] and val < list_quantile[1]:
                result.append('Average Orders')
            elif val >= list_quantile[1]:
                result.append('Most Orders')

ords_prods_merged['busiest_hours'] = result

ords_prods_merged['busiest_hours'].value_counts(dropna=False)
```
