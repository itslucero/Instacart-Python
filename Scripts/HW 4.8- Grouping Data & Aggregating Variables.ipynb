### 4. Follow the instructions in this Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Create an aggregate max column named max_order

ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merged.head()
```

```python
# Create flag

ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merged.loc[(ords_prods_merged['max_order'] > 10) & (ords_prods_merged['max_order'] <= 40), 'loyalty_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged.head()
```

```python
# Check frequency

ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

### 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

#### 
*If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

*If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create an aggregate average price column named average_price_spent

ords_prods_merged['average_price_spent'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# review df

ords_prods_merged.head(15)
```

```python
# Create a spending_flag with loc

ords_prods_merged.loc[ords_prods_merged['average_price_spent'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merged.loc[ords_prods_merged['average_price_spent'] >= 10, 'spending_flag'] = 'High spender' 
```

```python
ords_prods_merged.head()
```

```python
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

### 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

####

*If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

*If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

*If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create an aggregate median column named frequent customers

ords_prods_merged['frequent_customers'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merged.head(15)
```

```python
# Create a frequency_flag with loc

ords_prods_merged.loc[ords_prods_merged['frequent_customers'] > 20, 'frequency_flag'] ='Non-frequent customer'
ords_prods_merged.loc[(ords_prods_merged['frequent_customers'] > 10) & (ords_prods_merged['frequent_customers'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['frequent_customers'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Review dataframe

ords_prods_merged.head()
```

```python
ords_prods_merged['frequency_flag'].value_counts(dropna= False)
```
